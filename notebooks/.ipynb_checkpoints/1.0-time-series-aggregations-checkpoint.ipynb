# Time Series Aggregations
This challenge provides several datasets which contain a history of payments for each customer.  Over time, people change and therefore it would be interesting to take aggregations over only a subset of the payment history for a customer.  In this notebook, I explore the idea of aggregating over `k` last installments.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 

%matplotlib inline 
```

```python
path_to_data   = '../data/raw/'
app_train      = pd.read_csv(path_to_data + 'application_train.csv')
credit_card_df = pd.read_csv(path_to_data + 'credit_card_balance.csv')
```

```python
#test_id = np.random.choice(app_train['SK_ID_CURR'])
test_id = 128350
cc_df  = credit_card_df[credit_card_df['SK_ID_CURR'] == test_id]
print('Testing for SK_ID_CURR = %d, which has %d entries in the credit card dataset.' % (test_id, len(cc_df)))
```

```python
cc_df.head()
```

```python
cc_sorted = cc_df.sort_values(by=['MONTHS_BALANCE'], ascending=False)
cc_sorted.head(12)
```

```python
plt.plot(cc_sorted['MONTHS_BALANCE'], cc_sorted['AMT_BALANCE'])
plt.xlabel('Months')
plt.ylabel('Balance')
```

```python
def aggregate_over_k_installments(data, aggregations, k, title):
    max_installments = len(data)

    if k > max_installments:
        k = max_installments
    
    agg = data[:k].groupby('SK_ID_CURR').aggregate(aggregations)
    colname = lambda x, y: '%s_' % title + x + '_' + y.upper() + '_%d' % k
    new_cols = [colname(c[0],c[1]) for c in list(agg.columns)]
    agg.columns = new_cols
    
    return agg
```

```python
aggregations = {'AMT_BALANCE':['mean', 'var']}
```

```python
cc_agg_12 = aggregate_over_k_installments(cc_sorted, aggregations, 12, title='CC')
cc_agg_12.head()
```

```python
cc_agg_24 = aggregate_over_k_installments(cc_sorted, aggregations, 24, title='CC')
cc_agg_24.head()
```

```python
cc_agg = cc_agg_12.join(cc_agg_24)
cc_agg.head()
```

```python
periods = [6, 12, 18, 24]

cc_agg = aggregate_over_k_installments(cc_sorted, aggregations, periods[0], title='CC')
for k in periods[1:]:
    df = aggregate_over_k_installments(cc_sorted, aggregations, k, title='CC')
    cc_agg = cc_agg.join(df)
```

```python
cc_agg.head()
```

```python
customers = []

for sk_id in np.unique(credit_card_df['SK_ID_CURR'].values):
    data_subset = credit_card_df[credit_card_df['SK_ID_CURR'] == sk_id].copy()
    data_subset.sort_values(by='MONTHS_BALANCE', ascending=False, inplace=True)
    
    agg = aggregate_over_k_installments(data_subset, aggregations, periods[0], title='CC')
    for k in periods[1:]:
        
        if k < len(data_subset):
            df = aggregate_over_k_installments(data_subset, aggregations, k, title='CC')
            agg = agg.join(df)
        
    customers.append(agg)
```

```python
def last_agg(x, agg_fn, n):
    if len(x) < n:
        return np.nan 
    return agg_fn(x[:n])
```

```python
test_aggs = {'AMT_BALANCE':[lambda x: last_agg(x, np.mean, 12), lambda y: last_agg(y, np.var, 18)]}
cc_df.groupby('SK_ID_CURR').agg(test_aggs, title=['AGG1', 'AGG2'])
```
